```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/Users/raghunandanms/Documents/Personal/DS/IITK Datasets and other/used_cars_data.csv')
df
```

```python
df.isnull().sum()/len(df)*100
```

```python
df = df.drop(['S.No.'],axis=1)
df.info()
```

```python
df['Car_Age']=2024-df['Year']
df.head()
```

```python
df['Brand'] = df.Name.str.split().str.get(0)
df['Model'] = df.Name.str.split().str.get(1) + df.Name.str.split().str.get(2)
df[['Name', 'Brand', 'Model']]
```

```python
print(df.Brand.unique())
print(df.Brand.nunique())
```

```python
searchfor = ['Isuzu', 'ISUZU', 'Mini', 'Land']
df[df.Brand.str.contains('|'.join(searchfor))].head()
```

```python
df["Brand"].replace({"ISUZU": "Isuzu", "Mini": "Mini Cooper","Land":"Land Rover"}, inplace=True)
```

```python
fig, axes = plt.subplots(3, 2, figsize=(18, 18))
sns.countplot(ax=axes[0,0], x='Fuel_Type', data = df, color='blue', order=df['Fuel_Type'].value_counts().index)
sns.countplot(ax=axes[0,1], x='Transmission', data = df, color='blue', order=df['Transmission'].value_counts().index)
sns.countplot(ax=axes[1,0], x='Owner_Type', data = df, color='blue', order=df['Owner_Type'].value_counts().index)
sns.countplot(ax=axes[0,0], x='Fuel_Type', data = df, color='blue', order=df['Fuel_Type'].value_counts().index)
```

```python
num_col = df.select_dtypes(include=np.number).columns.tolist()
for cols in num_col:
    print(cols)
    print('Skew :', round(df[cols].skew(), 2))
    plt.figure(figsize = (15,4))
    plt.subplot(1,2,1)
    df[cols].hist(grid=False)
    plt.ylabel('Count')
    plt.subplot(1,2,2)
    sns.boxplot(x=df[cols])
    plt.show()
```
